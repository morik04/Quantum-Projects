Dinner Party Problem Problem

```python
#I want to invite Hamza, Arvand, Fares and Shervin to Dinner
#Hamza, Arvand and I are chill with everyone but Shervin and Fares do not like with each other
#I want at least two people to come
#So (Hamza or Arvand) and[(Shervin and not Fares) or (not Shervin and Fares)]

```

```python
from qiskit_aer import qasm_simulator, Aer
from qiskit_algorithms import Grover, AmplificationProblem
from qiskit.circuit.library import PhaseOracle
from qiskit.visualization import plot_histogram
from qiskit.primitives import Sampler
```

```python
log_expr = '(Hamza | Arvand) & ((Shervin & ~Fares) | (~Shervin & Fares))'
oracle = PhaseOracle(log_expr)
problem = AmplificationProblem(oracle)
```

```python
backend = Aer.get_backend('qasm_simulator')
sampler = Sampler()
```

```python
grover = Grover(sampler=sampler)
result = grover.amplify(problem)
```

```python
plot_histogram(result.circuit_results, title='Possible Combinations', bar_labels=True)
#Organized alphebetically so Arvand, Fares, Hamza, Shervin
```
